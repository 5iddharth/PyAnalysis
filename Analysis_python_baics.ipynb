```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("Sample-Spreadsheet-100-rows.csv")
df.head()
```

```python
df.columns = ["Col","Col1","Col2","Col3","Col4","Col5","Col6","Col7","Col8","Col9"]
df.drop_duplicates(subset = "Col2")
```

# Selecting Columns

```python
name = df["Col1"]
name
```

# Selecting Rows

```python
df[df["Col2"] == "Barry French"]
```

```python
df[df["Col"] == 2]
```

# Line Plot

```python
plt.plot(df.Col, df.Col2)
plt.show()
```

```python
plt.plot(df.Col3,df.Col4)
plt.title("Graph" , color = "Red" , fontsize = 20)
plt.style.use("seaborn")
plt.show()
```

# Scatter Plot

```python
plt.scatter(df.Col1, df.Col4)
plt.xlabel = "----"
plt.title("Scatter Plot" , color = "Pink")
plt.show()
```

#  Stacked Bar Chart

```python
plt.bar(df.Col2 , df.Col3 , label = "First")
plt.bar(df.Col2 , df.Col4 , bottom = df.Col3, label = "Second")
plt.bar(df.Col2 , df.Col5 , bottom = df.Col4, label = "Third")
plt.legend()
plt.show()
```

# Histogrm

```python
plt.hist(df.Col2 , bins= 1 , color="Orange")
plt.hist(df.Col3 , bins= 10 , color="Pink")
plt.show()
```
